```python
%%sql
create table bookings as select * from read_csv_auto('bookings.csv');
create table clients as select * from read_csv_auto('clients.csv');
```

```python
%%sql
with new_clients as (
    select
        client_id
    from clients
    where first_booking_date >= date '2025-01-01' and first_booking_date < date '2026-01-01'
)

select * from new_clients
```

```python
«Новый клиентов» нет. 
У всех клиентов в таблице clients first_booking_date приходится на 2024 год.
```
