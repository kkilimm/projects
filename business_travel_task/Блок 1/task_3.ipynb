```python
%%sql
create table bookings as select * from read_csv_auto('bookings.csv');
create table clients as select * from read_csv_auto('clients.csv');
```

```python
%%sql 
select
    subproduct,
    count(*) as total,
    sum(case when status = 'cancelled' then 1 else 0 end) as cancelled,
    round(1.0 * sum(case when status = 'cancelled' then 1 else 0 end) / count(*),4) as cancel_rate
from bookings
group by subproduct
order by cancel_rate desc;
```

```python
Субпродукта с заметно более высокой долей отмен не было выявлено.
У всех субпродуктов доля отмен находится в диапозоне примерно от 0.11 до 0.15 

Максимальная доля отмен у avia: 0.1525
Минимальная доля отмен у taxi: 0.1077
```
