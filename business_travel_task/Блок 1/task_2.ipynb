```python
%%sql
create table bookings as select * from read_csv_auto('bookings.csv');
create table clients as select * from read_csv_auto('clients.csv');
```

```python
%%sql 
with client_stats as (
    select
        client_id,
        round(sum(amount_rub),2) as gmv,
        count(*) as bookings_count,
        round(avg(amount_rub),2) as avg_check
    from bookings 
    where status = 'confirmed'
    group by client_id
    order by gmv desc
    limit 10
)

select 
    a.client_id as client_id, 
    gmv, 
    bookings_count, 
    avg_check, 
    tier, 
    industry
from client_stats as a left join clients as b
on a.client_id = b.client_id
order by gmv desc 
```
